Problem Statement:
The bank loan system aims to automate and optimize the process of assessing the creditworthiness of loan applicants. Using historical loan data and applicant attributes, the goal is to develop a machine learning model capable of accurately predicting whether a loan applicant is likely to repay the loan or default. By leveraging advanced analytics and predictive modeling techniques, the bank seeks to improve loan approval decisions, minimize default risk, and enhance overall operational efficiency.

# Train test split

```python
x = df.drop(['Loan_ID','Loan_Status'],axis=1)
y = df['Loan_Status']
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=11,stratify = y)
```

```python
x_train.isnull().sum()
```

```python
y_train.value_counts()
```

# Train Logistic regression

```python
log_clf = LogisticRegression()
log_clf.fit(x_train,y_train)
```

```python
y_pred_log = log_clf.predict(x_test)
cnf_matrix = confusion_matrix(y_test,y_pred_log)
print('confusion matrix :\n',cnf_matrix)

acc_score_LR = accuracy_score(y_test,y_pred_log)
print("Accuracy Score : ",acc_score_LR)

clf_report = classification_report(y_test , y_pred_log)
print('classification_report : \n',clf_report)
```

-------Interpretation of testing data-------

Precision for N(Loan did not pass): The model achieved a precision of 0.95 for the positive class, indicating that 95% of the instances predicted as positive were correct.
Precision for Y(Loan did pass): The model achieved a precision of 0.53 for the positive class, indicating that 53% of the instances predicted as positive were correct.


Recall for N: The recall for the positive class is 0.53, meaning that the model captured 53% of all actual positive instances.
Recall for Y: The recall for the positive class is 0.99, meaning that the model captured 99% of all actual positive instances.

F1-score for N: With an F1-score of 0.68, the model demonstrates a good balance between precision and recall.
F1-score for Y: With an F1-score of 0.90, the model demonstrates a good balance between precision and recall.

Support for N: The support value for the positive class is 38, indicating that there are 38 instances of the positive class in the dataset.
Support for Y: The support value for the positive class is 85, indicating that there are 85 instances of the positive class in the dataset.
